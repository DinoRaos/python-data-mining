# Webscraping
## static

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.thws.de/")
soup = BeautifulSoup(page.content, "html.parser")

print(soup)
```

```python
# Div element
soup.find("div")
```

```python
# Element by id
soup.find("div", id="panel-language")
```

```python
# Element by class
soup.find("div", class_="inner-col-align")
```

```python
# All Divs
results = soup.find_all("div")
for result in results:
    print(result)
    print()
```

```python
# All Divs & get the content
results = soup.find_all("div")
for result in results:
    print(result.text)
    print()
```

```python
# scraping tables
table = soup.find_all("table")
rows = table[0].find_all("tr")

for row in rows:
    cols = row.find_all("td")
    for col in cols:
        print(col.text)
```
