# Data visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Convert CSV to DF

```python
df = pd.read_csv("Hotel_Reviews.csv")
print(df)
```

### Plots

```python
df.plot()
plt.show()
```

```python
df.plot(y="Additional_Number_of_Scoring", x="Hotel_Name", kind="line", figsize=(20, 10), style='o-',
        markerfacecolor='w', color='b')
plt.suptitle("Test Title", fontsize=14, fontweight='bold')
plt.ylabel("Number of Scoring")
plt.xlabel("Hotel Name")
plt.show()
```

```python
print("min: " + str(df["Reviewer_Score"].min()) + "\nmax: " + str(df["Reviewer_Score"].max()) + "\nmean: " + str(
    df["Reviewer_Score"].mean()))
```

```python
df.plot(y="Reviewer_Score", kind="box", figsize=(20, 10), medianprops=dict(linewidth=3, color="red"))
plt.show()
```

```python
df.plot(kind="line", y="Total_Number_of_Reviews", x="Hotel_Name", color="yellow", figsize=(20, 10))
plt.show()
```

```python
top_reviewers = df.sort_values(by="Total_Number_of_Reviews_Reviewer_Has_Given", ascending=False).head(3)

top_reviewers.plot(kind="bar", x="Reviewer_Nationality", y="Total_Number_of_Reviews_Reviewer_Has_Given",
                   figsize=(20, 7), xlabel="Nationality", ylabel="Number of Reviews", color="green")
plt.show()
```
